```python
arr = []

for sid in SUBJ:
    
    print sid

    ospan_fta = np.nan
    if int(sid) in ospandf.pcode.values:
        ospan_fta = ospandf[ospandf.pcode==int(sid)]['fta.score'].values[0]
        ospan_math_acc = ospandf[ospandf.pcode==int(sid)]['processing.accuracy'].values[0]
        #ospan_fta = ospandf[ospandf.pcode==int(sid)]['max.span'].values[0]
    
    
    qd = questiondata(sid)
    
    if sid == '55':
        gender = 'male'
    elif sid == '65':
        gender = 'female'
    elif sid == '97':
        gender = 'male'
    elif sid == '107':
        gender = 'female'
    elif sid == '149':
        gender = 'female'
    else:
        gender = qd['gender']
    
    if 'age' in qd and qd['age']!='':
        age = int(qd['age'])
    elif sid == '149':
        age = 21
    elif sid == '144':
        age = 23
    elif sid == '55':
        age = 19
    else:
        age = np.nan

    
    dat = data(sid)
    activeitems = filter(lambda d: d[0]=='activeitems', dat)[0][1:-1]
    yokeditems = filter(lambda d: d[0]=='yokeditems', dat)[0][1:-1]
    
    
    if sid in ['1', 'michelexiong']:
        img_type_map = {'active': 'women', 'yoked': 'men'}
    elif sid == 'Arielle Little-retest':
        img_type_map = {'active': 'men', 'yoked': 'women'}
    else:
        cond_order = filter(lambda d: d[0]=='study_cond', dat)[0][1:]
        img_type_order = filter(lambda d: d[0]=='image_type', dat)[0][1:]
        img_type_map = {cond_order[i]: img_type_order[i] for i in range(2)}
    
    
    if cond_order[0]=='active':
        if img_type_order[0]=='men':
            counterbalance = 0
        else:
            counterbalance = 1
    else:
        if img_type_order[0]=='men':
            counterbalance = 2
        else:
            counterbalance = 3
    
    
    for cond in ['active', 'yoked']:
        
        for trial in range(N_TEST_TRIALS):
        
            testdata = filter(lambda d: d[0]=='test' and d[2]==cond and d[3]==trial, dat)      
            if len(testdata) > 0:
            
                t_onset = filter(lambda d: d[4]=='listen_for_response', testdata)[0][-1]

                testresp = filter(lambda d: d[4]=='response', testdata)[0]
                block = testresp[1]

                t_resp = testresp[-1]
                rt = t_resp - t_onset

                a = int(testresp[5].lstrip('id='))
                b = int(testresp[6].lstrip('id='))

                if cond == 'active':
                    rank_a = activeitems.index(a)
                    rank_b = activeitems.index(b)
                    dist = activeitems.index(b) - activeitems.index(a)
                else:
                    rank_a = yokeditems.index(a)
                    rank_b = yokeditems.index(b)
                    dist = yokeditems.index(b) - yokeditems.index(a)
                
                pair = pairs.index(list(np.sort([rank_a, rank_b])))
                
                endpoint = (rank_a in [0, 8]) or (rank_b in [0, 8])

                if dist == 1:
                    dist_bin = 1
                elif dist in [2,3]:
                    dist_bin = 2
                elif dist >= 4:
                    dist_bin = 3

                correct = testresp[8]
                arr.append([sid, gender, age, ospan_fta, ospan_math_acc, counterbalance, 0, cond, img_type_map[cond], block, trial, pair, a, b, rank_a, rank_b, dist, dist_bin, endpoint, correct, rt])

    if data('%s-retest' % sid) != None:
        dat = data('%s-retest' % sid)
        activeitems = filter(lambda d: d[0]=='activeitems', dat)[0][1:]
        yokeditems = filter(lambda d: d[0]=='yokeditems', dat)[0][1:]
        
        for cond in ['active', 'yoked']:

            for trial in range(N_TEST_TRIALS):

                testdata = filter(lambda d: d[0]=='test' and d[2]==cond and d[3]==trial, dat)            
                if len(testdata) > 0:
                
                    t_onset = filter(lambda d: d[4]=='listen_for_response', testdata)[0][-1]

                    testresp = filter(lambda d: d[4]=='response', testdata)[0]
                    block = testresp[1]

                    t_resp = testresp[-1]
                    rt = t_resp - t_onset

                    a = int(testresp[5].lstrip('id='))
                    b = int(testresp[6].lstrip('id='))

                    if cond == 'active':
                        rank_a = activeitems.index(a)
                        rank_b = activeitems.index(b)
                        dist = activeitems.index(b) - activeitems.index(a)
                    else:
                        rank_a = yokeditems.index(a)
                        rank_b = yokeditems.index(b)
                        dist = yokeditems.index(b) - yokeditems.index(a)
                        
                    pair = pairs.index(list(np.sort([rank_a, rank_b])))
                    endpoint = (rank_a in [0, 8]) or (rank_b in [0, 8])

                    if dist == 1:
                        dist_bin = 1
                    elif dist in [2,3]:
                        dist_bin = 2
                    elif dist >= 4:
                        dist_bin = 3


                    correct = testresp[8]
                    arr.append([sid, gender, age, ospan_fta, ospan_math_acc, counterbalance, 1, cond, img_type_map[cond], block, trial, pair, a, b, rank_a, rank_b, dist, dist_bin, endpoint, correct, rt])

testdf = pd.DataFrame(arr, columns=['sid', 'gender', 'age', 'ospan_fta', 'ospan_math_acc', 'counterbalance', 'session', 'cond', 'stim_type', 
                                    'block', 'trial', 'pair', 'lower_id', 'higher_id', 'lower_rank', 'higher_rank', 'distance', 'distance_bin',
                                    'endpoint', 'correct', 'rt'])
```

```python
# selections

arr = []
for sid in SUBJ:
    dat = data(sid)
    
    gender = testdf[testdf.sid==sid].gender.values[0]
    age = testdf[testdf.sid==sid].age.values[0]
    
    ospan_fta = np.nan
    if int(sid) in ospandf.pcode.values:
        ospan_fta = ospandf[ospandf.pcode==int(sid)]['fta.score'].values[0]


    cond_order = filter(lambda d: d[0]=='study_cond', dat)[0][1:]
    img_type_order = filter(lambda d: d[0]=='image_type', dat)[0][1:]
    img_type_map = {cond_order[i]: img_type_order[i] for i in range(2)}

    activeitems = filter(lambda d: d[0]=='activeitems', dat)[0][1:-1]
    yokeditems = filter(lambda d: d[0]=='yokeditems', dat)[0][1:-1]
    
    studydata = filter(lambda d: d[0]=='study' and d[2]=='active' and d[4]=='selection', dat)
    active_seq = np.array([activeitems.index(int(sd[6].lstrip('id='))) for sd in studydata])

    studydata = filter(lambda d: d[0]=='study' and d[2]=='yoked' and d[4]=='selection', dat)
    yoked_seq = np.array([yokeditems.index(int(sd[6].lstrip('id='))) for sd in studydata])
        
    for cond in ['active', 'yoked']:
        for trial in range(N_STUDY_TRIALS):
        
            studydata = filter(lambda d: d[0]=='study' and d[2]==cond and d[3]==trial, dat)
            block = studydata[0][1]
            
            options = filter(lambda d: d[4]=='option', studydata)
            options = [int(opt[6].lstrip('id=')) for opt in options]
            if trial == 0:
                overlap = np.nan
                overlap_feedback = np.nan
                near_distance = np.nan
            elif cond=='active':
                option_dist = [activeitems.index(opt) - active_seq[trial-1] for opt in options]
                near_distance = option_dist[np.argmin(np.abs(option_dist))]
                overlap_feedback = 1 in option_dist
                overlap = 1 in np.abs(option_dist)
            elif cond=='yoked':
                option_dist = [yokeditems.index(opt) - yoked_seq[trial-1] for opt in options]
                near_distance = option_dist[np.argmin(np.abs(option_dist))]
                overlap_feedback = 1 in option_dist
                overlap = 1 in np.abs(option_dist)
            
            if cond=='active':
                selected = active_seq[trial]
            else:
                selected = yoked_seq[trial]
            
            # response time
            t_onset = filter(lambda d: d[4]=='active_study_begin', studydata)[0][-1]
            t_selection = filter(lambda d: d[4]=='selection', studydata)[0][-1]
            rt = t_selection - t_onset
            
            # distance
            if trial == 0:
                dist_signed = np.nan
                dist = np.nan
            elif cond=='active':  
                dist_signed = active_seq[trial] - active_seq[trial-1]
                dist = np.abs(dist_signed)
            elif cond=='yoked':    
                dist_signed = yoked_seq[trial] - yoked_seq[trial-1]
                dist = np.abs(dist_signed)
                
            near = 1*(dist <= 2)
        
            arr.append([sid, gender, age, ospan_fta, cond, block, img_type_map[cond], trial, 
                        selected,
                        dist, dist_signed, near_distance, near, overlap, overlap_feedback, rt])

seldf = pd.DataFrame(arr, columns=['sid', 'gender', 'age', 'ospan_fta', 'cond', 'block', 'stim_type', 'trial', 
                                   'selected_ind',
                                   'distance', 'distance_signed', 'near_distance',
                                   'near', 'overlap', 'overlap_feedback', 'rt'])
```

```python
agg_seldf = seldf.groupby(['sid', 'block', 'cond']).apply(lambda d: pd.Series({'prop_near': d.near.mean(),
                                                                               'distance': d.distance.mean(),
                                                                               'n_overlap': np.sum(d.overlap),
                                                                               'n_overlap_feedback': np.sum(d.overlap_feedback),
                                                                               'prop_near_when_overlap': d[d.overlap==True].near.mean(),
                                                                               'prop_near_when_overlap_feedback': d[d.overlap_feedback==True].near.mean(),
                                                                               'prop_near_when_overlap_nonfeedback': d[(d.overlap==True) & (d.overlap_feedback==False)].near.mean(),
                                                                               'prop_near_when_nonoverlap': d[d.overlap==False].near.mean(),
                                                                               'prop_near_neg2': d[d.near_distance==-2].near.mean(),
                                                                               'prop_near_neg1': d[d.near_distance==-1].near.mean(),
                                                                               'prop_near_pos1': d[d.near_distance==1].near.mean(),
                                                                               'prop_near_pos2': d[d.near_distance==2].near.mean(),
                                                                               'med_rt': np.median(d.rt),
                                                                               'med_rt_overlap_feedback': np.median(d[d.overlap_feedback==True].rt),
                                                                               'med_rt_other': np.median(d[d.overlap_feedback==False].rt),
                                                                               })).reset_index()

# combine aggregate selection data with test data
cols = ['med_rt', 'med_rt_overlap_feedback', 'med_rt_other', 'distance', 'n_overlap', 'n_overlap_feedback', 'prop_near', 'prop_near_when_nonoverlap', 
        'prop_near_when_overlap_feedback', 'prop_near_when_overlap_nonfeedback', 'prop_near_when_overlap',
        'prop_near_neg2', 'prop_near_neg1', 'prop_near_pos1', 'prop_near_pos2']
for col in cols:
    testdf[('sel_'+col)] = np.nan

for sid in SUBJ:
    for cond in ['active', 'yoked']:
        
        sel = agg_seldf[(agg_seldf.sid==sid) & (agg_seldf.cond==cond)]
        
        for col in cols:
            scol = 'sel_' + col
            testdf.loc[(testdf.sid==sid) & (testdf.cond==cond),scol] = sel[col].values[0]        

testdf.to_csv('../data_proc/testdata.csv')
```
